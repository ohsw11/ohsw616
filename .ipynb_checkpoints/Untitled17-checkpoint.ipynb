```python
#1. Hello 버튼 만들기

import tkinter as tk #tkinter를 tk로 불러오기

def say_hello(): #say_hello 함수선언
    label.config(text="Hello, World!") #창에 "Hello, World!"를 뜨게 만들기

root = tk.Tk() #기본이 되는 윈도우 반환
button = tk.Button(root, text="Click Me", command=say_hello) #버튼을 만들고 버튼에 "Click Me"뜨게 만들기, 버튼을 누르면 say_hello 함수 실행
button.pack()
label = tk.Label(root, text="") #결과를 표시할 비어있는 lavel 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#2. 입력창에 이름을 입력하고 인사하기

import tkinter as tk #tkinter를 tk로 불러오기

def greet(): #greet 함수선언
    name = entry.get() #name을 작성할 구역 생성
    label.config(text=f"Hello, {name}") #활성화되면 창에 "Hello, {name}" 텍스트를 뛰음

root = tk.Tk() #기본이 되는 윈도우 반환
entry = tk.Entry(root)
entry.pack()
button = tk.Button(root, text="Greet", command=greet) #버튼을 만들고 버튼에 "Greet"뜨게 만들기, 버튼을 누르면 greet 함수 실행
button.pack()
label = tk.Label(root, text="") #결과를 표시할 비어있는 lavel 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#3. 간단한 계산기 만들기 (더하기)

import tkinter as tk #tkinter를 tk로 불러오기

def add(): #add 함수 선언
    result = int(entry1.get()) + int(entry2.get()) #entry1과 entry2 가입창의 텍스트를 문자열로 받은 다음 정수형으로 바꾼 뒤 서로 더
    label.config(text=f"Result: {result}") #화면에 "Result: {result}"가 뜨도록 수정

root = tk.Tk() #기본이 되는 윈도우 반환
entry1 = tk.Entry(root) #entry1 가입창 생성
entry1.pack()
entry2 = tk.Entry(root) #entry1 가입창 생성
entry2.pack()
button = tk.Button(root, text="Add", command=add) #"Add"라 적힌 버튼 만들기, 버튼을 누르면 add함수 실행
button.pack()
label = tk.Label(root, text="") #결과를 표시할 비어있는 lavel 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#4. 체크박스 선택 결과 출력하기

import tkinter as tk #tkinter를 tk로 불러오기

def show_selection(): #show_selection 함수 선언
    selections = [] #selections 리스트 생성
    if var1.get(): selections.append("Option 1") # var1 값이 True일때 Option1 선택
    if var2.get(): selections.append("Option 2") # var2 값이 True일때 Option2 선택
    label.config(text=", ".join(selections)) # 리스트 안의 요소를 ,로 나뉘어 표시

root = tk.Tk() #기본이 되는 윈도우 반환
var1 = tk.BooleanVar() #Var1을 BooleanVar로 생성
var2 = tk.BooleanVar() #Var2을 BooleanVar로 생성
tk.Checkbutton(root, text="Option 1", variable=var1).pack() #Option 1라고 적힌 체크박스 생성
tk.Checkbutton(root, text="Option 2", variable=var2).pack() #Option 2라고 적힌 체크박스 생성
tk.Button(root, text="Show", command=show_selection).pack() #Show라고 적힌 버튼 생성, 누르면 show_selection함수 실행
label = tk.Label(root, text="") #결과를 표시할 비어있는 lavel 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#5. 라디오 버튼으로 색상 선택하기

import tkinter as tk #tkinter를 tk로 불러오기

def change_color(): #change_color함수 선언
    root.configure(bg=color.get()) #선택된 버튼이 가진 값으로 뒷배경 표시

root = tk.Tk() #기본이 되는 윈도우 반환
color = tk.StringVar() #문자열 값 저장할 color 변수 생성
tk.Radiobutton(root, text="Red", variable=color, value="red", command=change_color).pack() #Red라고 적힌 라디오버튼 생성, color변수에 저장, 저장할 값 "red", 선택되면 change_color함수 실행
tk.Radiobutton(root, text="Blue", variable=color, value="blue", command=change_color).pack() #Blue라고 적힌 라디오버튼 생성, color변수에 저장, 저장할 값 "blue", 선택되면 change_color함수 실행
tk.Radiobutton(root, text="Green", variable=color, value="green", command=change_color).pack() #Green이라 적힌 라디오버튼 생성, color변수에 저장, 저장할 값 "green", 선택되면 change_color함수 실행
root.mainloop() #이벤트 처리
```

```python
#6. 간단한 폰트 크기 조절기 만들기

import tkinter as tk #tkinter를 tk로 불러오기

def resize(val): #resize함수 선언
    label.config(font=("Arial", int(val))) #폰트를 "Arial"로 설정, 슬라이더값으로 크기 설정

root = tk.Tk() #기본이 되는 윈도우 반환
scale = tk.Scale(root, from_=10, to=40, orient="horizontal", command=resize) #scale변수 생성, 10부터 40까지 값을 가짐, 수평방향로, 값이 변하면 resize함수 실행
scale.pack()
label = tk.Label(root, text="Resizable Text") #lavel에 "Resizable Text"뛰우기
label.pack()
root.mainloop() #이벤트 처리
```

```python
#7. 리스트박스로 항목 선택하기

import tkinter as tk #tkinter를 tk로 불러오기

def show_fruit(event): #show_fruit함수 선언
    selection = listbox.get(listbox.curselection()) #현재 선택된 항목의 텍스트를 가져옴
    label.config(text=f"Selected: {selection}") #Label에  "Selected: {selection}"표시

root = tk.Tk() #기본이 되는 윈도우 반환
listbox = tk.Listbox(root) #리스트박스 생성
for fruit in ["Apple", "Banana", "Cherry"]: #리스트 박스 끝에 "Apple", "Banana", "Cherry" 추가
    listbox.insert(tk.END, fruit)
listbox.pack()
listbox.bind("<<ListboxSelect>>", show_fruit) #항목 선택 이벤트 와 show_fruit함수 연결
label = tk.Label(root, text="") #결과를 표시할 비어있는 lavel 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#8. 메뉴바 추가하기

import tkinter as tk #tkinter를 tk로 불러오기

def show_message(msg): #show_message함수 선언
    label.config(text=msg) #label에 msg표시

root = tk.Tk() #기본이 되는 윈도우 반환
menu = tk.Menu(root) #메뉴바 생성
file_menu = tk.Menu(menu, tearoff=0) #분리되지 않는 file_menu 생성
file_menu.add_command(label="Open", command=lambda: show_message("Open clicked")) #"Open"항목 추가 클릭되면 show_message함수 실행
menu.add_cascade(label="File", menu=file_menu) #"File"을 메뉴바에 추가
help_menu = tk.Menu(menu, tearoff=0) #분리되지 않는 help_menu 생성
help_menu.add_command(label="About", command=lambda: show_message("About clicked")) #"About"항목 추가 클릭되면 show_message함수 실행
menu.add_cascade(label="Help", menu=help_menu) #"Help"를 메뉴바에 추가
root.config(menu=menu) #메뉴바를 창에 표시
label = tk.Label(root, text="") #결과를 표시할 비어있는 label 생성
label.pack()
root.mainloop() #이벤트 처리
```

```python
#9. 간단한 메모장 만들기

import tkinter as tk #tkinter를 tk로 불러오기

def save_file(): #save_file함수 선언
    with open("note.txt", "w") as f: #note.txt라는 파일을 쓰기 모드로 열기 
        f.write(text.get("1.0", tk.END)) #text 위젯의 내용을 가져와 파일에 씀

root = tk.Tk() #기본이 되는 윈도우 반환
text = tk.Text(root) #여러줄 입력이 가능한 텍스트 위젯 생성
text.pack()
button = tk.Button(root, text="Save", command=save_file) #save라고 적힌 버튼 생성, 누르면 save_file함수 실행
button.pack()
root.mainloop() #이벤트 처리
```

```python
#10. 실시간 시계 만들기

import tkinter as tk #tkinter를 tk로 불러오기
import time #time모듈 불러오기

def update_time(): #update_time함수 선언
    current = time.strftime("%H:%M:%S") #시/분/초의 정보를 문자엳로가지는 current변수 생성
    label.config(text=current) #label에 current 표시
    root.after(1000, update_time) #1초 후에 update_time함수 실행

root = tk.Tk() #기본이 되는 윈도우 반환
label = tk.Label(root, font=("Arial", 24)) # 24크기의 "Arial"폰트로 시간정보를 표시하는 label 생성
label.pack()
update_time() #update_time함수 실행
root.mainloop() #이벤트 처리
```
