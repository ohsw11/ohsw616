```python
#4.1.1.부호함수
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

def sgn(x):
    return np.sign(x)

x = np.linspace(-2, 4, 400)
y = sgn(x)

plt.plot(x, y)
plt.xlim(-2, 4)
plt.title("부호함수")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.grid(True)
plt.show()
```

```python
#4.1.1.단위계단함수
def heaviside_step(x):
    if isinstance(x, np.ndarray):
        return np.where(x >= 0, 1, 0)
    else:
        return 1.0 if x >= 0 else 0.0

x = np.linspace(-2, 4, 400)
y = heaviside_step(x)

plt.plot(x, y)
plt.xlim(-2, 4)
plt.title("단위계단함수")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.grid(True)
plt.show()
```

```python
#4.1.3
def logit(y):
    return np.log(y / (1 - y))

y = np.linspace(-2, 2, 10000)
plt.plot(y, logit(y))
plt.title("로지스틱함수의 역함수")
plt.xlabel("$y$")
plt.ylabel("logit(y)")
plt.grid(True)
plt.show()
```

```python
#4.2.5.2
import sympy

sympy.init_printing(use_latex='mathjax')

x, k  = sympy.symbols('x k')
f = sympy.log((x ** 2) -3*k)

sympy.simplify(sympy.diff(f, x))
```
